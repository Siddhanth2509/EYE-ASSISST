# EYE-ASSISST: Phase 2 - Frontend to Backend Integration

During Phase 2, the primary objective was replacing the simulated mock data in the React/Vite frontend with live API calls to the FastAPI backend, enabling real AI inference.

## Key Implementations:
- **Technician Portal Integration:** Modified `TechnicianDashboard` to transmit retinal fundus images using `FormData` (`multipart/form-data`) to the `/api/analyze` endpoint.
- **Response Handling:** Configured the frontend to parse the backend's JSON response, capturing:
  - `dr_binary` (Binary Diabetic Retinopathy detection)
  - `dr_severity` (5-class severity grading: Normal, Mild, Moderate, Severe, Proliferative)
  - `multi_disease` (Flags for Glaucoma, AMD, Cataracts)
  - `gradcam` (Base64-encoded explainability heatmap)
- **Image Heuristics:** Implemented a pre-upload heuristic check to ensure uploaded files exhibit a red-dominant color profile (typical of fundus images), rejecting invalid files (like scenery or faces) before backend processing.
